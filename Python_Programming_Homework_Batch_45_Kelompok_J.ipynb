```python
#3

list_id_category = ['c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9', 'c10']
list_bundling_category = list()

for kategori1 in list_id_category:
    category1 = kategori1
    for kategori2 in list_id_category:
        category2 = kategori2
        for kategori3 in list_id_category:
            category3 = kategori3

            # Lakukan proses pengurutan list
            list_sort = sorted([category1,category2,category3])

            # Lakukan proses filter dengan ketentuan yang berlaku
            if len(set(list_sort)) > 1 and list_sort not in list_bundling_category:
                list_bundling_category.append(list_sort)

# Tampilkan hasilnya
print('Terdapat', len(list_bundling_category), 'kemungkinan bundling category untuk keperluan campaign')

```

```python
#4
x = input('Ketik kata yang ingin di process:')
y = input('Ketik operasi yang diinginkan:')

def string_processor(x,y):
  def first_char(x):
    hasil = x[0]
    return hasil
  def last_char(x):
    hasil = x[len(x)-1]
    return hasil
  def num_chars(x):
    hasil = len(x)
    return hasil
  def num_vowels(x):
    hasil = 0
    vowels = ['a','i','u','e','o']
    for i in x:
      if i.lower() in vowels:
        hasil += 1
    return hasil

  if y == 'first_char':
    result = first_char(x)
  elif y == 'last_char':
    result = last_char(x)
  elif y == 'num_chars':
    result = num_chars(x)
  elif y == 'num_vowels':
    result = num_vowels(x)

  return result

print(string_processor(x,y))
```

```python
#6

import pandas as pd
import numpy as np

df = pd.read_excel('gdrive/MyDrive/superstore.xls')

def analysis(data, region, year=2017):
  data['Year'] = data['Order Date'].dt.year

  data = data[(data['Region'] == region)]
  data = data[(data['Year'] == year)]

  result = data.groupby(['Region','Year'], as_index = False).agg(
      total_sales = ('Sales', 'sum'),
      total_profit = ('Profit', 'sum')
  )
  return result

display(analysis(df,'South',2016))
display(analysis(df,'West',2017))
```
